```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
persons=pd.read_csv("https://gist.githubusercontent.com/javierIA/7d8564f02ae92bf722800465da2d0f7a/raw/6e0b018296ba1bafe2f20319c9f3e3a8573b151e/salarios.csv")
persons
```

```python
for i in range(5):
  print(persons.sample(frac=1/10),"\n")
```

```python
from sklearn.preprocessing import MinMaxScaler

#  Scaling all columns using MiMaxScaler()
persons=MinMaxScaler().fit_transform(persons)
persons
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import BaggingRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor

# Assigning columns "edad", "estudio" y "horas" to variable datos
datos=persons.T[:-2].T
# Assigning column "clase_salario" to variable clases
clases = persons.T[3]
# Assigning column "valor_salario" to variable valor
valores=persons.T[4]
valores

```

```python
from sklearn import neighbors

clasificador=KNeighborsClassifier(n_neighbors=3)
regresor=KNeighborsRegressor(n_neighbors=4)

numero_de_modelos=50

meta_class=BaggingClassifier(base_estimator=clasificador, n_estimators=numero_de_modelos, max_samples=0.3).fit(datos, clases)

meta_regression=BaggingRegressor(base_estimator=regresor, n_estimators=numero_de_modelos, max_samples=0.3).fit(datos, valores)
```

```python
#  Importing accuracy_socre and r2_score from sklearn.metrics
from sklearn.metrics import accuracy_score, r2_score

#  Calculating and printing the accuracy_score for the classification model
print("Class ", accuracy_score(clases, meta_class.predict(datos)))

#  Calculating and printing the r2_score for the regression model
print("Class ", r2_score(clases, meta_class.predict(datos)))
```
